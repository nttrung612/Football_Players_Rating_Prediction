```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
tmp = df.drop('nationality', axis=1)
```

```python
tmp.corr()['overall_rating'].sort_values(ascending=False)
```

```python
df['potential'] = df['potential'] - 5
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['overall_rating'], bins=30, alpha=0.5, label='Overall Rating')
plt.hist(df['potential'], bins=30, alpha=0.5, label='Potential')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Rating and Potential')
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['reactions'] = df['reactions'] + 2

plt.figure(figsize=(10, 6))
plt.hist(df['overall_rating'], bins=30, alpha=0.5, label='Overall Rating')
plt.hist(df['reactions'], bins=30, alpha=0.5, label='Reactions')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Rating and Reactions')
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['overall_rating'], bins=30, alpha=0.5, label='Overall Rating')
plt.hist(df['value_euro'], bins=30, alpha=0.5, label='Value')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Rating and Values')
plt.legend(loc='upper right')
plt.show()
```

```python
df.drop(['value_euro', 'wage_euro', 'release_clause_euro',
         'potential', 'composure', 'reactions'], axis=1, inplace=True)
```

```python
df.to_csv('data.csv', index=False)
```
